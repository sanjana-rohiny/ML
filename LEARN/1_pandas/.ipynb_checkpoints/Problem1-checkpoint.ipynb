```python
import numpy as np 
import pandas as pd
```

```python
# Import the data into Python environment as a Pandas DataFrame.
df = pd.read_csv('rainfall.csv')
df
```

```python
# Check for missing values, if any and drop the corresponding rows. 

# Dropping rows with missing values
df.dropna(inplace=True)
df
```

```python
# Find the district that gets the highest annual rainfall. 

index = df['ANNUAL'].idxmax()
print(index)
district_with_highest_rainfall = df.loc[index, 'DISTRICT']
print(district_with_highest_rainfall)
```

```python
# Display the top 5 states that get the highest annual rainfall.


# Sort the DataFrame by the 'ANNUAL' column in descending order
df_sorted = df.sort_values(by='ANNUAL', ascending=False)

# Display the top 5 states with the highest annual rainfall
top_5_states = df_sorted.head(5)[['STATE_UT_NAME', 'ANNUAL']]

print(top_5_states)
```

```python
# Drop the columns 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'. 

columns_to_drop = ['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec']
df.drop(columns=columns_to_drop, inplace=True)

# Display the updated column index
print(df.columns)
```

```python
# Display the state-wise mean rainfall for all the months using a pivot table. 

# Create a pivot table to calculate the mean rainfall for each month for each state
pivot_table_result = df.pivot_table(values=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN',
                                           'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'],
                                   index='STATE_UT_NAME',
                                   aggfunc='mean')

print(pivot_table_result)
```

```python
# Visualize pivot table. 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table_result, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5)
plt.title("State-wise Mean Rainfall for Each Month")
plt.xlabel("Month")
plt.ylabel("State")
plt.show()
```

```python
# Counting districts in each state

district_count_per_state = df['STATE_UT_NAME'].value_counts()

print(district_count_per_state)
```
