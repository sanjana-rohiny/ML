# Problem Statement:
# The engineers at XYZ Custom Cars want to infer the correlation between horsepower and mileage. 

# Solution:
# First, let us plot a scatter plot for horsepower and mileage.

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#importing the data
df = pd.read_csv('auto_mpg.csv')
df.info()    #to know the column names, missing values and their data type
```

```python
#replacing the null values in 'horsepower' with its mean value

# Convert 'horsepower' to numeric (if it's not already)
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Calculate the mean of 'horsepower' (excluding NaN values)
mean_horsepower = df['horsepower'].mean()

# Replace null values in 'horsepower' with its mean value
df['horsepower'].fillna(mean_horsepower, inplace=True)
```

```python
#visualizing correlation between mileage and horsepower
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[10,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
#plotting scatter plot
ax.scatter(df['horsepower'],df['mpg'])
ax.set_title("Scatter plot of horsepower and mileage",fontsize=16)
ax.set_ylabel('Mileage per gallon',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
fig.show()
```

# From the above image, it can be concluded that mileage and horsepower are negatively correlated to some extent. 

# Visualizing the correlation between mileage and horsepower based on the origin of the cars.

```python
#finding correlation between mileage and horsepower 
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[12,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Scatter plot of Horsepower and Mileage based on Origin",fontsize=16)
ax.set_ylabel('Mileage per gallon',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
#assigning different colors for each origin
origin_list = df["origin"].unique()
colors = plt.cm.jet(np.linspace(0, 1, len(origin_list))) 
# plotting based on the origin
for origin,color in zip(origin_list,colors):
    x = df[df["origin"] == origin]["horsepower"]
    y = df[df["origin"] == origin]["mpg"]
    plt.scatter(x,y,color = color,label=origin)
plt.legend()
```

#In the above Scatter plot different colors are assigned to different origins.
