```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2 * np.pi * x) * np.exp(-x)

plt.plot(x, y, 'b-', label='Line Plot')  # 'b-' sets the color to blue and line style to a solid line
plt.scatter(x, y, c='red', label='Highlighted Points', marker='o')  # 'c' sets the color to red, 'marker' sets the marker shape to a circle

# Adding the second crust point (x=2, y=0.5) with theta=60
x_second_crust = 2
y_second_crust = 0.5
theta_second_crust = 60
plt.scatter(x_second_crust, y_second_crust, c='green', label='Second Crust', marker='^')  # 'c' sets the color to green, 'marker' sets the marker shape to a triangle

plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot with Highlighted Points')
plt.grid(True)
plt.legend()
plt.show()
```
