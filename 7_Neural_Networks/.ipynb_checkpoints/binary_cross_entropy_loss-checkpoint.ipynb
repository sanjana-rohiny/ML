# binary_cross_entropy_loss

```python
import numpy as np
import matplotlib.pyplot as plt

def binary_cross_entropy_loss(y, y_hat):
    return -(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))

# Given data points
x_points = [1, 2, 3, 4, 5, 6, 7]
y_points = [1.5, 3.8, 6.7, 9.0, 11.2, 13.6, 16]

# Normalize y_points to be between 0 and 1
y_normalized = np.array(y_points) / np.max(y_points)

# Set true label y to 1 (as an example)
y_true_label = 1

# Calculate binary cross-entropy loss for each data point
loss_values = binary_cross_entropy_loss(y_true_label, y_normalized)

# Plot the binary cross-entropy loss
plt.figure(figsize=(8, 6))
plt.plot(x_points, loss_values, marker='o', linestyle='-', color='b')
plt.title('Binary Cross-Entropy Loss for Given Data Points')
plt.xlabel('X Points')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def binary_cross_entropy_loss(y, y_hat):
    return -(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))

# Generate a dataset of 100 points along a sigmoid curve
X = np.linspace(-7, 7, 100)
y_sigmoid = sigmoid(X)

# Set a true label (e.g., 1)
y_true_label = 1

# Calculate binary cross-entropy loss
loss_values = binary_cross_entropy_loss(y_true_label, y_sigmoid)

# Plot the sigmoid curve and binary cross-entropy loss
plt.figure(figsize=(12, 6))

#plt.subplot(1, 2, 1)
#plt.plot(X, y_sigmoid, marker='o', linestyle='-', color='b')
#plt.title('Sigmoid Function')
#plt.xlabel('X')
#plt.ylabel('Sigmoid(X)')
#plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(X, loss_values, marker='o', linestyle='-', color='r')
plt.title('Binary Cross-Entropy Loss')
plt.xlabel('X')
plt.ylabel('Loss')
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Heaviside step function

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the Heaviside step function
def heaviside_step(x):
    return np.where(x >= 0, 1, 0)

# Generate x values
x_values = np.linspace(-5, 5, 1000)

# Calculate y values based on the Heaviside step function
y_values = heaviside_step(x_values)

# Plot the Heaviside step function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Heaviside Step Function', color='blue')
plt.title('Heaviside Step Function')
plt.xlabel('x')
plt.ylabel('H(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
