```python
# Packages for analysis
import pandas as pd
import numpy as np
from sklearn import svm

# Packages for visuals
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=1.2)

# Allows charts to appear in the notebook
%matplotlib inline

# Pickle package
import pickle
```

```python
# Read in muffin and cupcake ingredient data
recipes = pd.read_csv('recipes_muffins_cupcakes.csv')
recipes
```

# Plot two ingredients - seaborn plot


```python
# take only two attributes, sugar and flour, since   we plot on two axis
sns.lmplot(x='Flour', y='Sugar', data=recipes, hue='Type',
           palette='Set1', fit_reg=False, scatter_kws={"s": 70});
```

# seaborn plot shows  a clear separation in blue andred spot
#   Hence Sugar and  Flour  can be used to classify...

```python
# Specify inputs for the model
# ingredients = recipes[['Flour', 'Milk', 'Sugar', 'Butter', 'Egg', 'Baking Powder', 'Vanilla', 'Salt']].as_matrix()

# sugar and  Flour  can be used to classify...
ingredients = recipes[['Flour','Sugar']].values

#  specify output value... 0 -> indicate Muffin and 1-> indicate Cupcake
type_label = np.where(recipes['Type']=='Muffin', 0, 1)
print(ingredients)
```

# Extract  features  or  attributes or independent variables

```python

# Feature names
recipe_features = recipes.columns.values[1:].tolist()
recipe_features
```

```python
# Fit the SVM model

model = svm.SVC(kernel='linear')

#  fit(features, targetvalues) -> here only  two features
model.fit(ingredients, type_label)
```

```python
# Get the separating hyperplane
w = model.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(30, 60)
yy = a * xx - (model.intercept_[0]) / w[1]

# Plot the parallels to the separating hyperplane that pass through the support vectors
b = model.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])
b = model.support_vectors_[-1]
yy_up = a * xx + (b[1] - a * b[0])
```

```python
# Plot the hyperplane
sns.lmplot(x='Flour', y='Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black');

```

```python
# Look at the margins and support vectors
sns.lmplot(x='Flour', y='Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            s=80, facecolors='none');
```

```python
# Create a function to guess when a recipe is a muffin or a cupcake
def muffin_or_cupcake(flour, sugar):
    if(model.predict([[flour, sugar]]))==0:
        print('You\'re looking at a muffin recipe!')
    else:
        print('You\'re looking at a cupcake recipe!')
```

```python
# Predict if 50 parts flour and 20 parts sugar
muffin_or_cupcake(50, 20)
```

```python
# Plot the point to visually see where the point lies
sns.lmplot(x='Flour', y='Sugar', data=recipes, hue='Type', palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black')
plt.plot(50, 20, 'yo', markersize='9');
```

```python
# Predict if 40 parts flour and 20 parts sugar
muffin_or_cupcake(40,20)
```

# Create  dictionary named muffin_cupcake_dict. It contains three key-value pairs:

    Key: 'muffin_cupcake_model', Value: model - The value associated with this key is referenced by the variable model.

    Key: 'muffin_cupcake_features', Value: ['Flour','Sugar'] - The value associated with this key is a list containing the features 'Flour' and 'Sugar'.

    Key: 'all_features', Value: recipe_features - The value associated with this key is referenced by the variable recipe_features.
    
    For instance, one could retrieve the model and its features from the dictionary to perform predictions on new recipes 

```python
muffin_cupcake_dict = {'muffin_cupcake_model': model, 'muffin_cupcake_features': ['Flour','Sugar'], 'all_features': recipe_features}

```

```python
muffin_cupcake_dict
```

```python
# Pickle
pickle.dump(muffin_cupcake_dict, open("muffin_cupcake_dict.p", "wb"))

```

```python
# S = String
pickle.dumps(muffin_cupcake_dict)
```

#  Retrieve the model and its features from the dictionary 
# to perform predictions on new recipes 

```python
def muffin_or_cupcake(flour, sugar):
    model = muffin_cupcake_dict['muffin_cupcake_model']

    # Accessing the list of features from the dictionary
    muffin_cupcake_features = muffin_cupcake_dict['muffin_cupcake_features']

    # New recipe with specific values for 'Flour' and 'Sugar'
    new_recipe = {'Flour': 40, 'Sugar': 20}

    # Prepare the input features for the model
    input_features = [new_recipe[feature] for feature in muffin_cupcake_features]

    # Make the prediction using the model
    prediction = model.predict([input_features])
    return int(prediction[0])
```

# Example usage:


```python

# Example usage:
flour_amount = 40
sugar_amount = 20
prediction = muffin_or_cupcake(flour_amount, sugar_amount)

if prediction == 0:
    print("You're looking at a muffin recipe!")
else:
    print("You're looking at a cupcake recipe!")
```
