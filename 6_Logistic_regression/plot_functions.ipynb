

# Cos Function

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -2π to 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Calculate y values for f(x) = cos(x)
y = np.cos(x)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='f(x) = cos(x)')
plt.title('Plot of f(x) = cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = (1/2) * x^2
def f(x):
    return (1/2) * x**2

# Generate x values
x = np.linspace(-10, 10, 400)  # Create 400 equally spaced points from -10 to 10

# Calculate corresponding y values
y = f(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = (1/2) * x^2', color='b')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = (1/2) * x^2')
plt.grid(True)
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f1(x):
    return (1/2) * x**2

def f2(x):
    return x

# Generate x values
x = np.linspace(-2, 2, 500)

# Calculate corresponding y values for both functions
y1 = f1(x)
y2 = f2(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='f(x) = (1/2) * x^2', color='b')
plt.plot(x, y2, label='f(x) = x', color='r', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = (1/2) * x^2 and f(x) = x')
plt.grid(True)
plt.legend()
plt.show()
```

# function f(x)=x3−6x2+9x+3f(x)=x3−6x2+9x+3:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema

# Define the function f(x)
def custom_function(x):
    return x**4 - 6*x**2 + 9*x + 3

# Generate x values
x = np.linspace(-3, 3, 1000)

# Calculate corresponding y values
y = custom_function(x)

# Find local minima and maxima
local_minima_indices = argrelextrema(y, np.less)
local_maxima_indices = argrelextrema(y, np.greater)

# Calculate the second derivative of the function
y2 = np.gradient(y, x)  # Second derivative using numpy's gradient function

# Find saddle points (where the second derivative is zero)
saddle_point_indices = np.where(y2 == 0)[0]

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x)')
plt.scatter(x[local_minima_indices], y[local_minima_indices], color='r', marker='v', label='Local Minima')
plt.scatter(x[local_maxima_indices], y[local_maxima_indices], color='g', marker='^', label='Local Maxima')
plt.scatter(x[saddle_point_indices], y[saddle_point_indices], color='m', marker='s', label='Saddle Points')

# Find and label the global minimum
global_minimum_x = x[np.argmin(y)]
global_minimum_y = min(y)
plt.scatter(global_minimum_x, global_minimum_y, color='b', marker='o', label='Global Minimum')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function with Global and Local Extrema, and Saddle Points')
plt.grid(True)
plt.legend()
plt.show()

# Output the coordinates of the extrema and saddle points
print("Global Minimum: ({:.2f}, {:.2f})".format(global_minimum_x, global_minimum_y))
print("Local Minima: ", [(x[i], y[i]) for i in local_minima_indices[0]])
print("Local Maxima: ", [(x[i], y[i]) for i in local_maxima_indices[0]])
print("Saddle Points: ", [(x[i], y[i]) for i in saddle_point_indices])
```
